```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im1=cv2.imread('1.jpeg')
im1=cv2.resize(im1,(250,250))
im2=cv2.imread('2.jpeg')
im2=cv2.resize(im2,(250,250))
```

```python
im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
im2=cv2.cvtColor(im2,cv2.COLOR_BGR2RGB)
```

```python
sift=cv2.SIFT_create()
```

```python
keypoints_1,descriptors_1=sift.detectAndCompute(im1,None)
keypoints_2,descriptors_2=sift.detectAndCompute(im2,None)
```

```python
bf=cv2.BFMatcher(cv2.NORM_L1,crossCheck=True)
matches=bf.match(descriptors_1,descriptors_2)
matches=sorted(matches,key=lambda x:x.distance)
```

```python
im3=cv2.drawMatches(im1,keypoints_1,im2,keypoints_2,matches[:10],im2,flags=2)
plt.imshow(im3)
plt.show()
```
