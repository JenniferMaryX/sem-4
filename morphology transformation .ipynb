### morphology transformation and image segmentation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## DILATION

```python
im1=cv2.imread("26.jpeg",0)
im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
_,mask=cv2.threshold(im1,220,255,cv2.THRESH_BINARY)
kernal=np.ones((29,29),np.uint8)
dilation=cv2.dilate(mask,kernal,iterations=1)
erode=cv2.erode(mask,kernal,iterations=1)
```

```python
titles=['img','mask','dilation','erode']
images=[im1,mask,dilation,erode]
```

```python
for i in range (len(images)):
    plt.subplot(1,4,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

## morphological open

```python
Open=cv2.morphologyEx(im1,cv2.MORPH_OPEN,kernal)
```

```python
cv2.imshow('img',im1)
cv2.imshow('morph open',Open)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## morpholocial close

```python
 Close=cv2.morphologyEx(im1,cv2.MORPH_CLOSE,kernal)
```

```python
cv2.imshow('img',im1)
cv2.imshow('morph CLOSE',Close)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## morphological gradient

```python
gradient=cv2.morphologyEx(im1,cv2.MORPH_GRADIENT,kernal)
```

```python
cv2.imshow('img',im1)
cv2.imshow('morph gradient',gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
titles=['img','dilation','erode','morp open','morph close','morph gradient']
images=[im1,dilation,erode,Open,Close,gradient]
```

```python
for i in range (len(images)):
    plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

## morphological top hat

```python
topHat=cv2.morphologyEx(im1,cv2.MORPH_TOPHAT,kernal)
```

```python
cv2.imshow('img',im1)
cv2.imshow('morph top hat',topHat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## morphological black hat(opp of top hat)

```python
blackHat=cv2.morphologyEx(im1,cv2.MORPH_BLACKHAT,kernal)
```

```python
cv2.imshow('img',im1)
cv2.imshow('morph black hat',blackHat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### morph open -erosion followed by dilation
#### morph close - dilation followed by erosion
#### morph gradient - erosion,dilation difference of both
#### morph top hat - opening and diff. of original and opening
#### morph black hat - opp of top hat
