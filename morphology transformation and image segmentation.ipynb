### morphology transformation and image segmentation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## DILATION

```python
im1=cv2.imread("26.jpeg",0)
im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
_,mask=cv2.threshold(im1,220,255,cv2.THRESH_BINARY)
kernal=np.ones((3,3),np.uint8)
dilation=cv2.dilate(mask,kernal,iterations=1)
erode=cv2.erode(mask,kernal,iterations=1)
```

```python
titles=['img','mask','dilation','erode']
images=[im1,mask,dilation,erode]
```

```python
for i in range (len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
