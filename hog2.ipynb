```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import matplotlib.pyplot as plt
```

```python
img=imread('25.jpeg')
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

## resize image

```python
resize_img=resize(img,(128*10,64*10))
plt.axis("off")
plt.imshow(resize_img)
plt.show()
print(resize_img.shape)
```

```python
fd,hog_img=hog(img,orientations=10,pixels_per_cell=(2,2),cells_per_block=(4,4),visualize=True,channel_axis=2)
print(fd.shape)
print(hog_img.shape)
plt.axis("off")
plt.imshow(hog_img,cmap='hot')
plt.show()
```

```python
fd,hog_img=hog(img,orientations=10,pixels_per_cell=(2,2),cells_per_block=(4,4),visualize=True,channel_axis=2)
print(fd.shape)
print(hog_img.shape)
plt.axis("off")
plt.imshow(hog_img,cmap='grey')
plt.show()
```
